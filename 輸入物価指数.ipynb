````python
# ライブラリのインポート（Colabには最初から入ってるで）
import pandas as pd
import matplotlib.pyplot as plt

# 日本語フォントの設定（グラフ用）
plt.rcParams['font.sans-serif'] = ['DejaVu Sans']
````

````python
# FREDから日本の輸入物価指数を取得
series_id = "JPNTOT"
url = f"https://fred.stlouisfed.org/graph/fredgraph.csv?id={series_id}"

# データ読み込み（parse_datesは後で指定）
df = pd.read_csv(url)

# カラム名を確認
print("Column names:", df.columns.tolist())

# 日付カラムを変換
df.columns = ['Date', 'Import_Price_Index']
df['Date'] = pd.to_datetime(df['Date'])

print("Data shape:", df.shape)
print("\nFirst 5 rows:")
print(df.head())
print("\nLast 5 rows:")
print(df.tail())

# 基本統計量
print("\nBasic Statistics:")
print(df.describe())

# グラフ表示
plt.figure(figsize=(12, 6))
plt.plot(df['Date'], df['Import_Price_Index'], linewidth=2)
plt.title('Import Price Index for Japan (All Industries)', fontsize=14, fontweight='bold')
plt.xlabel('Date', fontsize=12)
plt.ylabel('Index', fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````
